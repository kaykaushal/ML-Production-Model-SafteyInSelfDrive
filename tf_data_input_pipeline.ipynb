```python
from __future__ import absolute_import, division, print_function, unicode_literals
import os
import glob
import tensorflow as tf
import pathlib
tf.__version__
```

### Load Text Data into the framework

```python
DATA_DIR = "/tf/data/Downloads/2011_09_26/2011_09_26_drive_0001_sync/"
txt_paths = os.listdir(DATA_DIR)
#img_path = os.listdir(DATA_DIR+'/image_02/data')
```

```python
#text_data_path = 'oxts/data/'
def text_dataset(data_dir, text_data_dir, **args):
    paths = glob.glob(data_dir + text_data_dir + '*') #list of data file path
    data = tf.data.TextLineDataset(paths) # read data from each file in paths
    return data
```

```python
#Run function to make collection of text file from directories
text_data = text_dataset(DATA_DIR,'oxts/data/')
```

```python
for line in text_data.take(3):
      print(line.numpy())
```
